## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
### Create a column that calculates the subtotal for each line using the unit_price and the qty
# 	unit_price	qty	    line_subtotal
# 0	1096.80	    105	     115164.00
# 1	24.95	    21	     523.95

# line_subtotal = (unit_price) x (qty)

df["line_subtotal"] = df['unit_price'] * df['qty']
df[['unit_price', 'qty', 'line_subtotal']].head(2)
# Create a column that calculates the subtotal for each line using the unit_price and the qty. ---------------------------------> (6 points)
```

```python
### Create a column for shipping price.
### Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

#       unit_price	unit_weight	qty	    total_weight	shipping_price
# 0	    1096.80	    7.50	    105	    787.50	        5512.50
# 1	    24.95	    1.49	    21	    31.29	        312.90
# 2	    13.52	    1.68	    39	    65.52	        458.64

# unit_weight
# shipping_price
# MAKE → total_weight
# MAKE → shipping_price

# shipping_price = total_weight x ($7 if over 50)
#                                 (10 if 50 or under) → ELSE
df["total_weight"] = df['unit_weight'] * df['qty']
df["shipping_price"] = df\
    .apply(lambda x: (7 * x['total_weight']) if x['total_weight'] > 50 else (10 * x['total_weight']),
           axis=1)
df[['unit_price','unit_weight', 'qty','total_weight', 'shipping_price']].head(3)
# Create a column for shipping price. Assume a shipping price of $7 per pound for orders over 50 pounds
# and $10 per pound for items 50 pounds or under. -------------------------------------------------------------------------> (6 points)


```

```python
### Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
# 	    line_subtotal	shipping_price	line_price
# 0	    115164.00	    5512.50	        131839.08
# 1	    523.95	        312.90	        914.26
# 2	    527.28	        458.64	        1077.12

# line_subtotal
# shipping_price
# MAKE → line_price

df["line_price"] = round((df['line_subtotal'] + df['shipping_price']) * 1.0925,2)
df.head(3)
#df[['line_subtotal','shipping_price', 'line_price']].head(3)
#Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%.  ----------------------->  (6 points)

```
